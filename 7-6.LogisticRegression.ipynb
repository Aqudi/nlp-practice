## 6절 텐서플로우를 사용한 로지스틱 회귀 실습

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers


def makeY(x):
    return False if x < 10 else True


# 숫자 10 이상인 경우에는 1, 10 미만인 경우에는 0을 부여
X = np.arange(-50, 51)
makeYFun = np.vectorize(makeY)
y = makeYFun(X)

model = Sequential()

# 활성화 함수로 sigmoid 사용
model.add(Dense(1, input_dim=1, activation='sigmoid'))
sgd = optimizers.SGD(lr=0.01)

# 이진 분류 시에 cross entropy 함수를 사용한다면
# 손실 함수로 binary cross entropy 사용
model.compile(optimizer=sgd, loss='binary_crossentropy',
              metrics=['binary_accuracy'])
model.fit(X, y, epochs=2000)
```

```python
Xy = np.array(list(zip(X, y)))
Xy.sort(axis=0)
sortedX = Xy[:, 0]
sortedY = Xy[:, 1]
plt.plot(sortedX, model.predict(sortedX), 'b', sortedX, sortedY, 'k.')
```

```python
print(model.predict([-100, 4.5, 5.9, 7.5, 8.3, 9.7]), "\n")
print(model.predict([10.000001, 10.5, 11.2, 15.5, 500]))
```
